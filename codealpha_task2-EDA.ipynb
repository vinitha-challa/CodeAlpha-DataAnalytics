Missing values

```python
df.isnull().sum()
```

Duplicate Records

```python
df.duplicated().sum()
```

No duplicate record found

Total Sales

```python
df["Sales"].sum()
```

Sales by Category

```python
df.groupby("Category")["Sales"].sum()
```

observation: Technology category generated the highest sales

Profit by Category

```python
df.groupby("Category")["Profit"].sum()
```

observation: Technology category generated the highest proifit

Sales by Region

```python
df.groupby("Region")["Sales"].sum()
```

observation: West region recorded the highest sales.

Sales by segment

```python
df.groupby("Segment")["Sales"].sum()
```

observation: Consumer segment contributed the highest sales.

Top 10 Products

```python
df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

observation: Canon imageCLASS 2200 Advanced Copier was the top-selling product.

Monthly Sales Analysis

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Month"] = df["Order Date"].dt.month

df.groupby("Month")["Sales"].sum()
```

observation:November recorded the highest sales.
