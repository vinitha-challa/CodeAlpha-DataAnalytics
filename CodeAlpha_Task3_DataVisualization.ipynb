Code Cell

```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

## Observation

Technology category generated the highest sales.

Profit by category

```python
profit_category = df.groupby("Category")["Profit"].sum()

profit_category.plot(kind="bar")

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.show()
```

Observation: Technology category generated the highest profit.

Sales by Region

```python
region_sales = df.groupby("Region")["Sales"].sum()

region_sales.plot(kind="bar")

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

Observation: West region recorded the highest sales.

Sales by Segment

```python
segment_sales = df.groupby("Segment")["Sales"].sum()

segment_sales.plot(kind="pie", autopct="%1.1f%%")

plt.title("Sales by Segment")
plt.ylabel("")

plt.show()
```

Observation: Consumer segment contributed the highest sales.

Sales Distribution

```python
plt.hist(df["Sales"], bins=30)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

Observation: Most sales are concentrated in lower-value transactions.

Sales vs Profit

```python
plt.scatter(df["Sales"], df["Profit"])

plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()
```

Observation: Higher sales do not always result in higher profit.


Discount vs Profit

```python
plt.scatter(df["Discount"], df["Profit"])

plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")

plt.show()
```

Observation: Profit generally decreases as discount increases.

Monthly Sales Trend

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Month"] = df["Order Date"].dt.month

monthly_sales = df.groupby("Month")["Sales"].sum()

monthly_sales.plot(kind="line")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
Observatio: November recorded the highest sales.
```

# Conclusion

- Technology category generated the highest sales and profit.
- West region contributed the highest sales.
- Consumer segment contributed the highest sales.
- Higher discounts negatively affected profitability.
- November was the highest sales month.
- Some high-sales orders still resulted in losses.
